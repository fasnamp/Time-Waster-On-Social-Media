# Time-Wasters on Social Media
This dataset contains user behavior and demographic data related to social media usage . It captures various aspects of user activity, including time spent, engagement metrics, and platforms used. Additionally, it includes demographic and contextual details such as age, gender, income, device type, and connection type.

The dataset provides a valuable resource for analyzing social media consumption patterns, factors influencing productivity loss, and behavioral trends across different demographic segments.

## Key Features


* UserID: Unique identifier assigned to each user.
* Age: The user's age.
* Gender: The user's gender (e.g., male, female, non-binary).
* Location: Geographic location of the user.
* Income: The user's income level.
* Debt: Amount of debt the user has.
* Owns Property: Indicates whether the user owns property.
* Profession: The user's occupation or job.
* Demographics: Statistical data about the user (e.g., age, gender, income).
* Platform: The platform the user is using (e.g., website, mobile app).
* Total Time Spent: The total time the user spends on the platform.
* Number of Sessions: The number of times the user has logged into the platform.
* Video ID: Unique identifier for a video.
* Video Category: The category or genre of the video.
* Video Length: Duration of the video.
* Engagement: User interaction with the video (e.g., likes, comments, shares).
* Importance Score: A score indicating how important the video is to the user.
* Time Spent On Video: The amount of time the user spends watching a video.
* Number of Videos Watched: The total number of videos watched by the user.
* Scroll Rate: The rate at which the user scrolls through content.
* Frequency: How often the user engages with the platform.
* Productivity Loss: The impact of platform usage on the user's productivity.
* Satisfaction: The user's satisfaction level with the platform or content.
* Watch Reason: The reason why the user is watching a video (e.g., entertainment, education).
* Device Type: The type of device the user is using (e.g., smartphone, tablet, desktop).
* OS: The operating system of the user's device (e.g., iOS, Android, Windows).
* Watch Time: The time of day when the user watches videos.
* Self Control: The user's ability to control their usage of the platform.
* Addiction Level: The user's level of dependency on the platform.
* Current Activity: What the user is doing while watching the video.
* Connection Type: The type of internet connection the user has (e.g., Wi-Fi, cellular).

```python
plt.figure(figsize=(10,5))
df.boxplot()
plt.xticks(rotation=90)
plt.show()
```

```python
df['Gender'].value_counts().plot(kind='pie',autopct='%1.0f%%',shadow=True,colors=['red','yellow','green'])
plt.title('Gender',color='black',fontsize=20)
plt.savefig('Gender.png',dpi=300)
plt.show()
```

#### The pie chart shows that 51% of the dataset identifies as Male, 32% as Female, and 16% as Other. This highlights a majority of Male entries in the gender distribution.


```python
plt.figure(figsize=(10,6))
plt.title('Analysis of Reasons',color='black',fontsize=20,fontweight='bold')
sns.countplot(x='Watch Reason',data=df,hue='Gender',palette=['yellow','blue','grey'])
plt.savefig('Reasons.png',dpi=300)
plt.show()
```

#### The most common reason to watch social media is a Habit,especially males.Entertainment and Boredom are also significant reasons

```python
df['Age']
```

```python
age_addiction=df.groupby('Age')['Addiction Level'].mean()
```

```python
plt.figure(figsize=(10,6))
plt.title('Age-based addiction',color='black',fontsize=20,fontweight='bold')
plt.bar(age_addiction.index,age_addiction.values,color='brown')
plt.xlabel('Age',color='blue',fontsize=16,fontweight='bold')
plt.ylabel('Addiction Level',color='blue',fontsize=16,fontweight='bold')
plt.savefig('Addiction',dpi=300)
plt.show()
```

#### It shows the addiction levels fluctuates across ages with noticeable spikes in middle age ranges

```python
plt.figure(figsize=(10,5))
plt.title('Correlation Heatmap of Social media time-Waters',color='black',fontsize=20,fontweight='bold')
numer=df.select_dtypes('number')
sns.heatmap(numer.corr(),annot=True)
plt.show()
```

#### The heatmap shows that addiction is strongly linked to behavioral factors like time spent, videos watched, and self-control, while demographic factors such as age and income have little impact

```python

plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x="Total Time Spent", y="Addiction Level", hue="Gender", palette="deep")
plt.title("Total time spend vs Addiction Level", fontsize=16, fontweight='bold')
plt.xlabel("Total Time Spent", fontsize=12, fontweight='bold')
plt.ylabel("Addiction Level", fontsize=12, fontweight='bold')
plt.show()
```

#### The scatterplot shows that higher total time spent on social media generally corresponds to higher addiction levels, with little difference across genders.


```python
plt.figure(figsize=(10,6))
sns.lineplot(data=df, x="Age", y="Watch Time", hue="Gender", marker="o")
plt.title("Watch Time vs Age by Gender", fontsize=16, fontweight='bold')
plt.xlabel("Age", fontsize=12, fontweight='bold')
plt.ylabel("Watch Time", fontsize=12, fontweight='bold')
plt.show()
```

#### Watch time Varies by age but shows no major differences across genders.

```python
df['Profession']
```

```python
stacked_data=df.groupby(['Profession','Gender'])['Self Control'].mean().unstack(fill_value=0)
plt.figure(figsize=(10,6))
stacked_data.plot(kind="bar",stacked=True,figsize=(12,6),colormap="tab20")
plt.title("Average Self Control by Profession and Gender", fontsize=16, fontweight='bold')
plt.xlabel("Profession")
plt.ylabel("Average Self Control")
plt.show()
```

#### Self-control scores are steady across professions, with no major gender differences.

```python
plt.figure(figsize=(10,6))
plt.title('Engagement across Locations',fontsize=20,color='red',fontweight='bold')
sns.violinplot( data=df,x="Location",y="Engagement",hue="Platform")
plt.xlabel('Location')
plt.ylabel('Engagement')
plt.show()
```

#### Engagement levels are broadly consistent across platforms and locations, though countries like Pakistan and India show more variability in engagement compared to others.

```python
plt.figure(figsize=(10,6))
plt.title('Average Scroll Rate by Platform',color='blue',fontsize=20,fontweight='bold')
plt.xticks(rotation=45)
plt.xlabel('scroll rate',color='b')
plt.ylabel('platform',color='b')
plt.barh(df['Platform'],df['Scroll Rate'],color='grey')
plt.savefig('top 10 city.png',dpi=300)
plt.show()
```

#### Scrolling is a universal behavior across social media platforms, with little variation between them.
